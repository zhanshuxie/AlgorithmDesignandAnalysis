 # 部分背包问题

```python
def greedy_fractional_knapsack(weights, values, capacity):

    items = list(zip(weights, values))
    # 按单位重量价值(价值/重量)降序排列
    items.sort(key=lambda x: x[1] / x[0], reverse=True)

    total_value = 0

    remaining_capacity = capacity

    while remaining_capacity > 0:
    # 背包未装满，遍历
        for weight, value in items:
            # 若物品可全部装入，全装
            if remaining_capacity >= weight:
                remaining_capacity -= weight
                total_value += value
                print(f"全装物品（重量{weight}kg，价值{value}元）→ 累计价值{total_value}元，剩余容量{remaining_capacity}kg")
                if remaining_capacity == 0:
                    print("背包已满！")
                    break
            # 若装不下，取部分装入
            else:
                fraction = remaining_capacity / weight
                total_value += fraction * value
                print(f"装物品部分（重量{weight}kg的{fraction:.2f}，价值{value*fraction:.2f}元）→ 累计价值{total_value:.2f}元，背包已满")
                remaining_capacity = 0
                break
    return total_value
```

```python
weights = [3, 2, 4]
values = [15, 12, 16]
c = 5
result = greedy_fractional_knapsack(weights, values, c)
print(f"最终最大价值：{result:.2f}元")
```

# 凸多边形划分

```python
#
def convex_polygon_triangulation(n):

    if n < 3:
        return 0

    # 初始化dp表格
    dp = [[0] * n for _ in range(n)]
    # 填充边界条件
    for i in range(n):
        if i + 1 < n:
            # j = i + 1: 2个顶点 空多边形
            dp[i][i + 1] = 1
        if i + 2 < n:
            # j = i + 2: 3个顶点 三角形
            dp[i][i + 2] = 1

    for length in range(4, n + 1):
        # length:多边形边数
        for i in range(n + 1 - length):
            # j < n -> i + length - 1 < n -> i < n + 1 - length
            j = i + length - 1
            # 当i = 0 length = 4时 j = 3
            # 遍历所有中间状态
            for k in range(i + 1, j):
                dp[i][j] += dp[i][k] * dp[k][j]

    return dp[0][n-1]
```

```python
# 测试案例（验证卡特兰数对应关系）
if __name__ == "__main__":
    # 卡特兰数序列（C₀=1, C₁=1, C₂=2, C₃=5, C₄=14, C₅=42...）
    # n边凸多边形方案数 = Cₙ₋₂
    test_cases = [3, 4, 5, 6, 7]  # 3边（三角形）、4边（四边形）、5边（五边形）等
    for n in test_cases:
        count = convex_polygon_triangulation(n)
        print(f"{n} 边凸多边形的三角划分方案数：{count}（对应卡特兰数 Cₙ₋₂ = C_{n-2}）")
```

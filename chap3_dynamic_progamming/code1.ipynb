# 0-1 背包问题

![](./../source/figures/ch3_code1.png)

n=3 个物品，w=[2,3,4]，v=[3,4,5]，背包容量 C=5

```python
# 0-1背包问题
def zero_one_knapsack_two_dimensions(weights, values, capacity):
    """

    :param weights: 物品重量列表(w[i]为第i个物品的重量,索引从0开始)
    :param values: 物品价值列表(v[i]为第i个物品的价值,索引从0开始)
    :param capacity: 背包最大容量
    :return:
    """
    n = len(weights)

    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # 遍历所有物品、容量
    for i in range(1, n + 1):
        # i 为可装入物品最大编号
        for j in range(1, capacity + 1):
            # j 为当前背包容量

            # 第i个物品的重量、价值
            w_i = weights[i - 1]
            v_i = values[i - 1]
            # 如果装不下当前物品
            if w_i > j:
                dp[i][j] = dp[i - 1][j]
            # 如果装下当前物品
            else:
                dp[i][j] = max(dp[i - 1][j - w_i] + v_i, dp[i - 1][j])

    return dp[n][capacity]
```

```python
w = [2, 3, 4]
v = [3, 4, 5]
C = 5

res = zero_one_knapsack_two_dimensions(w, v, C)
print(f"背包内的最大价值为{res}")
```
